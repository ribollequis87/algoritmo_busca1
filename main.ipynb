```python
from aigyminsper.search.SearchAlgorithms import BuscaLargura
from aigyminsper.search.SearchAlgorithms import BuscaProfundidade
from aigyminsper.search.SearchAlgorithms import BuscaProfundidadeIterativa
from aigyminsper.search.Graph import State
from datetime import datetime
import pandas as pd

class SumOne(State):

    def __init__(self, n, op, g):
        self.operator = op
        self.number = n
        self.goal = g

    def successors(self):
        sucessors = []
        if self.number < self.goal:
            sucessors.append(SumOne(self.number+1, "+1 ", self.goal))
            sucessors.append(SumOne(self.number+2, "+2 ", self.goal))
        return sucessors

    def is_goal(self):
        if self.goal == self.number:
            return True
        return False

    def description(self):
        return "Este é um agente simples que sabe somar 1 e 2"

    def cost(self):
        return 1

    def env(self):
        return self.number
```

```python
def largura(numero):
    objetivo = numero
    state = SumOne(0, '', objetivo)
    algorithm = BuscaLargura()
    start_time = datetime.now()
    result = algorithm.search(state)
    end_time = datetime.now()
    time = end_time - start_time
    if result != None:
        return "Achou", time
        # print(result.show_path())
    else:
        return "Não Achou", time
    
resultado_largura = {}
for i in range(1, 51):
    conclusion, time = largura(i)
    resultado_largura[f"{i}  {conclusion}"] = time
```

```python
df_largura = pd.DataFrame(resultado_largura.items(), columns=['Chave', 'Valor'])
df_largura
```

```python
df_profundidade_dez = pd.DataFrame(resultado_profundidade_dez.items(), columns=['Solução', 'Tempo'])
df_profundidade_dez
```
